### Ver si mejoran las métricas con XGBoost

```python
from pycaret.regression import setup, compare_models, pull, stack_models, blend_models, evaluate_model

# Setup inicial
exp = setup(
    data=df_pycaret,
    target="PRECIO M2",
    session_id=123,
    normalize=True,
    remove_multicollinearity=True,
    multicollinearity_threshold=0.8,
    imputation_type="simple",
    numeric_imputation="median",
    categorical_imputation="mode",
    use_gpu=True,
    fold_strategy="timeseries",
    data_split_shuffle=False,   
    fold_shuffle=False
)

# Comparar modelos (ahora debería aparecer también XGBoost si se instaló bien)
top_models = compare_models(n_select=5, sort="R2")
```

```python
# Stacking: combina varios modelos en un meta-modelo
stacked_model = stack_models(
    estimator_list=top_models,   # lista de mejores modelos
    meta_model=None,             # si lo dejas en None, usa LinearRegression por defecto
    fold=5,                      # validación cruzada
    optimize="R2"
)

# Evaluar el modelo apilado
evaluate_model(stacked_model)
```

```python
blended_model = blend_models(
    estimator_list=top_models,
    fold=5,
    optimize="R2"
)

evaluate_model(blended_model)
```

Se aprecia una mejora significativa al aplicar OHE.

La comparativa de algoritmos mostró que los modelos lineales ofrecen el mejor rendimiento en este conjunto de datos.
El modelo Orthogonal Matching Pursuit (OMP) alcanzó un R² de 0.9183 con un RMSE de 201 €/m², siendo el más preciso.

Otros modelos lineales como Lasso, LassoLars, Ridge y Bayesian Ridge obtuvieron un desempeño ligeramente inferior, con valores de R² entre 0.71 y 0.76 y RMSE entre 280 y 305 €/m².

En contraste, los métodos de ensamblado basados en árboles (Random Forest, Gradient Boosting, XGBoost, LightGBM y Extra Trees) mostraron un rendimiento más bajo, con R² ≤ 0.78 y RMSE por encima de 250 €/m².

Finalmente, modelos más complejos como Elastic Net, KNN, Passive Aggressive o Huber Regressor tampoco lograron mejorar el desempeño, quedándose con R² entre 0.30 y 0.60 y errores significativamente mayores.

Estos resultados confirman que, en este dataset, la relación entre las variables explicativas y el precio por m² se ajusta mejor a un modelo de tipo lineal con codificación adecuada de variables categóricas, que a estructuras no lineales más complejas como los ensamblados.

🔝 Mejores modelos según R² y error

Orthogonal Matching Pursuit (omp) →

R² = 0.9183, el más alto de todos los modelos.

RMSE ≈ 201 €/m², MAE ≈ 156 €/m² → errores significativamente más bajos que el resto.
✅ Es el mejor modelo: combina precisión y simplicidad, mostrando que la relación entre las variables y el precio/m² es casi lineal.

Gradient Boosting Regressor (gbr) →

R² = 0.7847, bastante inferior a OMP.

RMSE ≈ 257 €/m², MAE ≈ 204 €/m².
Un modelo aceptable, pero con más error y menor capacidad explicativa.

Lasso, LassoLars, Ridge, Bayesian Ridge →

R² entre 0.71 y 0.76.

RMSE entre 280 y 305 €/m².
Modelos lineales regulares, más interpretables, pero claramente menos precisos que OMP.

📊 Modelos de ensamblado (XGBoost, Random Forest, LightGBM, Extra Trees)

R² entre 0.65 y 0.72, lo que indica que no capturan bien la estructura de los datos en este escenario.

RMSE entre 298 y 369 €/m², bastante superiores a los modelos lineales ajustados con OHE.
⚠️ Contrario a lo esperado, aquí los árboles no aportan ventaja.

⚖️ Otros modelos

Elastic Net, Passive Aggressive, Huber, KNN, AdaBoost, Decision Tree: rendimiento intermedio o bajo, con R² ≤ 0.60 y errores altos.

Linear Regression y Least Angle Regression → se desestabilizan y dan resultados inválidos (valores negativos de R²).

Dummy Regressor → R² ≈ 0, básicamente predice la media.

⏱️ Tiempos de entrenamiento (TT Sec)

Todos los modelos entrenan en menos de 1 segundo.

OMP es extremadamente rápido y a la vez el más preciso.

✅ Conclusión práctica
Los datos, tras imputación y one hot encoding, muestran un patrón altamente lineal.
El mejor modelo es Orthogonal Matching Pursuit, que logra el equilibrio óptimo entre precisión, interpretabilidad y eficiencia.
Los modelos de ensamblado, que en otros contextos suelen dominar, aquí no superan a los lineales y muestran un ajuste bastante peor.
