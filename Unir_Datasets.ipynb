## 5. Precio viviendas distritos

```python
precio_distritos = pd.read_excel(r"C:\Users\evahr\Downloads\TFM-idealista\Precio_viviendas_distritos.xlsx")

precio_distritos.tail(10)
```

```python
import pandas as pd
import numpy as np
import re
import unicodedata

# --- normalizador de distritos ---
def norm_dist(x):
    if pd.isna(x): return x
    x = str(x).strip()
    x = ''.join(ch for ch in unicodedata.normalize('NFKD', x) if not unicodedata.combining(ch))
    return ' '.join(x.split())

# --- parser EU: quita miles (solo cuando corresponde) y convierte coma decimal ---
_thousands_re = re.compile(r'(?<=\d)[\.\s](?=\d{3}(?:\D|$))')
def parse_eu(val):
    if pd.isna(val): return np.nan
    s = str(val).strip().replace('\xa0',' ')
    s = s.replace('%','')                      
    s = _thousands_re.sub('', s)               
    s = s.replace(',', '.')                    
    return pd.to_numeric(s, errors='coerce')

# ===================== 1) precio_distritos: wide -> long =====================
meses = ['Enero','Febrero','Marzo','Abril','Mayo','Junio',
         'Julio','Agosto','Septiembre','Octubre','Noviembre','Diciembre']
mes_map = {m:i+1 for i,m in enumerate(meses)}

pdw = precio_distritos.copy().rename(columns={'Año':'AÑO','Distrito':'DISTRITO'})
pdw['DISTRITO'] = pdw['DISTRITO'].map(norm_dist)
pdw = pdw[pdw['DISTRITO'] != 'ciudad de madrid']

months_present = [c for c in meses if c in pdw.columns]
pdw = pdw[['AÑO','DISTRITO'] + months_present]

# limpiar y convertir respetando decimales (19,37 -> 19.37)
pdw.loc[:, months_present] = pdw.loc[:, months_present].replace(
    {'..': np.nan, '—': np.nan, '–': np.nan, '-': np.nan}
).infer_objects(copy=False)
pdw.loc[:, months_present] = pdw.loc[:, months_present].applymap(parse_eu)

precio_long = (pdw.melt(id_vars=['AÑO','DISTRITO'], value_vars=months_present,
                        var_name='MES_NOMBRE', value_name='PRECIO_M2_DIST')
                 .dropna(subset=['PRECIO_M2_DIST']))
precio_long['MES_NUM'] = precio_long['MES_NOMBRE'].map(mes_map).astype('Int64')
precio_long = (precio_long.sort_values(['AÑO','DISTRITO','MES_NUM'])
                          .drop_duplicates(['AÑO','DISTRITO','MES_NUM'], keep='last')
                          [['AÑO','MES_NUM','DISTRITO','PRECIO_M2_DIST']])

# ===================== 2) merged: normaliza claves =====================
m = merged.copy()
m['AÑO']      = pd.to_numeric(m['AÑO'], errors='coerce').astype('Int64')
m['MES_NUM']  = pd.to_numeric(m['MES_NUM'], errors='coerce').astype('Int64')
m['DISTRITO'] = m['DISTRITO'].map(norm_dist)

# ===================== 3) Unión por AÑO + MES_NUM + DISTRITO =====================
final = pd.merge(
    m, precio_long,
    on=['AÑO','MES_NUM','DISTRITO'],
    how='left',
    validate='many_to_one'
)

# final tiene PRECIO_M2_DIST correctamente como 19.37, 20.43, etc.

final.head(30)
```

```python
import unicodedata

# --- 1) Mes nombre -> número (robusto con acentos/espacios) ---
def mes_a_num(x: str) -> int:
    x = str(x).strip().lower()
    x = ''.join(ch for ch in unicodedata.normalize('NFKD', x) if not unicodedata.combining(ch))
    mapa = {
        'enero':1,'febrero':2,'marzo':3,'abril':4,'mayo':5,'junio':6,
        'julio':7,'agosto':8,'septiembre':9,'setiembre':9,
        'octubre':10,'noviembre':11,'diciembre':12
    }
    return mapa.get(x, pd.NA)

# --- 2) Prepara euríbor (1 fila por año-mes) ---
eur = euribor.copy()
eur['MES_NUM'] = eur['Mes'].apply(mes_a_num).astype('Int64')
eur = eur.rename(columns={'Año':'AÑO', 'Valor':'EURIBOR'})


# --- 3) Merge many-to-one con tu DF mensual (no eliminamos duplicados de 'final') ---
final_con_euribor = pd.merge(
    final,
    eur[['AÑO','MES_NUM','EURIBOR']],
    on=['AÑO','MES_NUM'],
    how='left',            # 'inner' u 'outer' si prefieres
    validate='many_to_one' # muchas filas en 'final' por mes, 1 en 'eur'
)

# listo: 'final_con_euribor' tiene la columna EURIBOR replicada para cada distrito en ese año-mes

final_con_euribor.head(20)
```

```python
import pandas as pd
import numpy as np
import unicodedata
import re

# --- helpers ---
def tri_a_num(s: str):
    s = str(s).strip().lower()
    s = ''.join(ch for ch in unicodedata.normalize('NFKD', s)
                if not unicodedata.combining(ch))
    for ch in s:
        if ch.isdigit():
            return int(ch)
    return pd.NA

# Convierte número europeo "157,23" o "1.234,56" -> 157.23 / 1234.56
_thousands_re = re.compile(r'(?<=\d)[\.\s](?=\d{3}(?:\D|$))')
def parse_eu(val):
    if pd.isna(val): return np.nan
    s = str(val).strip().replace('\xa0',' ')
    s = _thousands_re.sub('', s)   
    s = s.replace(',', '.')
    return pd.to_numeric(s, errors='coerce')

# ========= 1) Prepara precios_vivienda (Año, Trimestre, Dato) =========
pv = precios_vivienda.copy()

# Normaliza nombres si hiciera falta
pv = pv.rename(columns={'Año':'AÑO','Trimestre':'TRIMESTRE','Dato':'DATO'})

pv['AÑO'] = pd.to_numeric(pv['AÑO'], errors='coerce').astype('Int64')
pv['TRI'] = pv['TRIMESTRE'].apply(tri_a_num).astype('Int64')
pv['DATO'] = parse_eu(pv['DATO']) if pv['DATO'].dtype == 'O' else pv['DATO']

# Mapea cada trimestre a sus meses y “explota” a formato mensual
q_to_months = {1:[1,2,3], 2:[4,5,6], 3:[7,8,9], 4:[10,11,12]}
pv['MES_NUM'] = pv['TRI'].map(q_to_months)
pv_long = (pv[['AÑO','MES_NUM','DATO']]
           .dropna(subset=['AÑO','MES_NUM'])
           .explode('MES_NUM', ignore_index=True)
           .rename(columns={'DATO':'INDICE_PRECIO'})
          )

# Si hubiera duplicados por AÑO+MES (no debería), nos quedamos con el último
pv_long = (pv_long.sort_values(['AÑO','MES_NUM'])
                  .drop_duplicates(['AÑO','MES_NUM'], keep='last'))

# ========= 2) Une con tu DF mensual por AÑO+MES_NUM (sin eliminar filas) =========
base = final_con_transmisiones.copy()
base['AÑO'] = pd.to_numeric(base['AÑO'], errors='coerce').astype('Int64')
base['MES_NUM'] = pd.to_numeric(base['MES_NUM'], errors='coerce').astype('Int64')

final_con_precios = pd.merge(
    base,
    pv_long,                                  
    on=['AÑO','MES_NUM'],
    how='left',
    validate='many_to_one'
)

final_con_precios.head()
```

```python

# --- helpers ---
def tri_a_num(x):
    s = str(x).lower()
    m = re.search(r'[1-4]', s)
    return int(m.group()) if m else pd.NA

q_to_months = {1:[1,2,3], 2:[4,5,6], 3:[7,8,9], 4:[10,11,12]}

_thousands_re = re.compile(r'(?<=\d)[\.\s](?=\d{3}(?:\D|$))')
def to_num_es(v):
    if pd.isna(v): return np.nan
    s = str(v).replace('\xa0',' ').strip()
    s = _thousands_re.sub('', s).replace(',', '.')
    return pd.to_numeric(s, errors='coerce')

# 1) Izquierdo (mensual)
left = final_con_precio_medio_metrocuadrado_ccmm.copy()
left['AÑO'] = pd.to_numeric(left['AÑO'], errors='coerce').astype('Int64')
left['MES_NUM'] = pd.to_numeric(left['MES_NUM'], errors='coerce').astype('Int64')
n_before = len(left)

# 2) Derecho (trimestral -> mensual)
val = valor_transacciones_suelo.copy()
val.columns = [c.replace('\u00A0',' ').strip() for c in val.columns]
val = val.rename(columns={'VALOR TRANSACCIONES SUELO':'VALOR_TRANSACCIONES_SUELO'})
val['AÑO'] = pd.to_numeric(val['AÑO'], errors='coerce').astype('Int64')
val['TRIM_NUM'] = val['TRIMESTRE'].apply(tri_a_num).astype('Int64')
if val['VALOR_TRANSACCIONES_SUELO'].dtype == 'O':
    val['VALOR_TRANSACCIONES_SUELO'] = val['VALOR_TRANSACCIONES_SUELO'].map(to_num_es)

val['MES_NUM'] = val['TRIM_NUM'].map(q_to_months)
val_m = (val[['AÑO','MES_NUM','VALOR_TRANSACCIONES_SUELO']]
         .dropna(subset=['AÑO','MES_NUM'])
         .explode('MES_NUM', ignore_index=True))

# si hubiera duplicados por AÑO+MES, suma
val_key = (val_m.groupby(['AÑO','MES_NUM'], as_index=False)['VALOR_TRANSACCIONES_SUELO']
                .sum())

# 3) LEFT merge → dataset_final
dataset_final = left.merge(val_key, on=['AÑO','MES_NUM'], how='left', validate='many_to_one')
assert len(dataset_final) == n_before  

# (opcional) combos de AÑO-MES de la base sin match:
# no_match = (left[['AÑO','MES_NUM']].drop_duplicates()
#             .merge(val_key[['AÑO','MES_NUM']], on=['AÑO','MES_NUM'], how='left', indicator=True)
#             .query("_merge=='left_only'").drop(columns=['_merge']))


dataset_final.head()
```
